## 1. Request 라이브러리로 response 가져와보기
또 들어가기 전에 배경지식 타임을 가집시다


request 와 response는 메시지 형식을 지닙니다.<br>
### Request
메소드-요청헤더-/공백/-본문으로 구성됩니당.<br> 
여기서 주목할건<br>
#### 메소드: 클라이언트가 서버에 요청하는 방식<br>
GET, POST, DELETE등이 있으나 보통 보안상의 이유로 웹서버가 앞 두개만 허용합니다. 둘 다 **웹서버에 자원을 요청하는** 메소드입니다. 두 방식의 차이점을 알아두면 아주 죠습니다.<br>
- **GET**: 데이터를 url뒤에 추가하어 전달합니다. 입력값이 많지 않은 경우, 간단한 데이터를 빠르게 처리할 수 있습니다. 그러나 데이터양에 제한이 있고, 보안 중요한 곳에 쓰면 위험합니다 데이터 노출~ 만천하에 공개~<br>
ex) 검색어, 페이지번호. www.yeojin.com/search?q=사람&p=2
- **POST**: 데이터를 url과 별도로 전송합니다. 보통 데이터를 서버로 제출하여 추가/수정하기 위해 데이터를 전송하는 방식입니다. 최소한의 보안 유지 효과가 있으며 get보다 많은 데이터를 전송할 수 있습니다.<br>
ex) 로그인(id, pw가 url에 뜨면 큰일나겟조)

### Response
상태코드-응답헤더-/공백/-본문으로 구성됩니다.<br>
여기서 볼것은
#### 상태코드: 클라이언트의 요청에 대한 서버의 처리 결과
200번대(2XX)는 성공, <br>
300번대(3XX)는 리다이렉션 완료, <br>
400번대(4XX)는 클라이언트 오류, <br>
500번대(5XX)는 서버 오류를 의미합니다.

## 2.BeautifulSoup으로 파싱하기
BeautifulSoup은 bs4 라이브러리에 있는 함수로, HTML을 ^예쁘게^ 파싱해줍니다. 

```python
from bs4 import BeautifulSoup
```

```python
# response.ok를 통해 응답이 잘 도착했는지 확인합니다.
response = requests.get("https://naver.com")

if response.ok:
    print("Request accepted.")
else:
    print("Request failed.")
    print(response.status_code, response.reason)
```

```python
# BeautifulSoup으로 html 태그를 인식하는 객체를 만들 수 있습니다.
# lxml은 html을 인식해주는 친구! html.parser라는 친구도 html을 인식해주지만 느립니다. lxml은 C로 짜여져 있어서 빠름.

soup = BeautifulSoup(response.text, "lxml")
```

### 2) find(), find_all() 메소드

```python
# find 메소드를 사용할 경우 tag 이름과 속성 값까지 사용하여 원하는 태그의 값을 찾을 수 있습니다.
# 태그 이름이 span이며 class 속성값으로 "an_txt"를 갖는 태그를 가져오는 코드입니다.
soup.find("span", {"class" : "an_txt"})
```

```python
# 가져온 값에서 텍스트만 분리하고 싶을 경우에는 get_text() 메소드를 사용합니다.
soup.find("span", {"class" : "an_txt"}).get_text()
```

```python
# find_all 메소드를 사용하면 조건에 맞는 태그를 리스트에 담아 모두 가져옵니다.

soup.find_all("span", {"class" : "an_txt"})
```

```python
# 만약 text만 담은 리스트를 가져오고 싶다면, 리스트 변수를 새로 설정하여 담아줍니다.

category_list = soup.find_all("span", {"class" : "an_txt"})
category_list = [raw_text.get_text() for raw_text in category_list]
```

```python
response = requests.get("https://sports.news.naver.com/volleyball/index.nhn")

if response.ok:
    print("Request accepted.")
else:
    print("Request failed.")
    print(response.status_code, response.reason)
```

### 3. 크롤링한 파일 저장하기

```python
# 네이버 스포츠 - 야구 카테고리의 뉴스 기사 제목과 내용을 크롤링하여 파일에 저장하는 코드입니다.
# 각 기사의 내용을 저장하기 위해서는 기사 링크가 필요한데, 
# 기사 링크는 https://sports.news.naver.com/kbaseball/news/read.nhn?oid={신문사 id}&aid={기사 id}로 이루어져 있습니다.
# 따라서 우선 기사 제목, oid, aid를 가져오겠습니다.
```

```python
import pandas as pd
```

```python
url = "https://sports.news.naver.com/kbaseball/news/list.nhn"


# 저는 2018년 3월 6일의 첫번째 페이지만 가져오지만 몇 년 치를 가져와야 한다면 for문으로 date / page의 값만 바꾸면 되겠죠!
date = "20180306"
page = "1"

params = {
    "date" : date,
    "isphoto" : "N",
    "type" : "latest",
    "page" : page
}

response = requests.get(url, params = params)
```

```python
news_json = response.json()
raw_news_list = news_json['list']
news_dict = [{'title' : news['title'], 'oid' : news['oid'], 'aid' : news['aid']} for news in raw_news_list]
```

```python
news_result = []

for news in news_dict:
    
    title = news["title"]
    oid = news["oid"]
    aid = news["aid"]
    
    params = {
        "oid" : oid,
        "aid" : aid
    }
    
    response = requests.get("https://sports.news.naver.com/kbaseball/news/read.nhn", params = params)
    soup = BeautifulSoup(response.text, "lxml")
    content = soup.select_one("#newsEndContents").get_text()
    
    news_result.append([title, oid, aid, content])
```

```python
content_df = pd.DataFrame(news_result, columns=["title" , "oid", "aid", "content"])
```

```python
content_df.to_csv("news_result.csv", index=False) # 1열에 0, 1, 2, ... 인덱스 붙는게 싫다면 index = False하시면 됩니다.
```
